```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Importation of the necessary modules for the project
```

```python
df = pd.read_csv(r"C:\Users\hp\Desktop\csv_folders\fifa21_raw_data.csv", low_memory=False)
#The CSV file was read from files

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
#This is used to display the entire rows and columns

sample_df = df.sample(n=500, random_state=42)
#Random Sampling Technique was applied because the DataSet is has about 18,000 rows and its much to work with.

sample_df
#This printed out the CSV table

#Reading of the CSV file
```

```python
sample_df = sample_df.drop(columns=["playerUrl", "photoUrl"])
# Removed the irrelevant columns

sample_df = sample_df.reset_index()

sample_df["Team & Contract"] = df["Team & Contract"].str.replace("\n", "", regex = False).str.replace("~", "-", regex = False).str.replace(r'(\D)(\d{4})', r'\1 \2', regex=True)
sample_df["Hits"] = df["Hits"].str.replace("\n", "")

sample_df["Loan Date End"] = df["Loan Date End"].fillna("-")



sample_df['Weight'] = sample_df['Weight'].str.extract(r'(\d+\.?\d*)').astype(float)


sample_df["Weight"] = (sample_df["Weight"] * 0.453592).round(0).astype(int)

#A function to convert from feet and inches to meters
def height_to_meters(height):
    feet, inches = map(int, height.replace('"', "").split("'"))
    return round(feet * 0.3048 + inches * 0.0254, 2)

sample_df["Height"] = sample_df["Height"].apply(height_to_meters) 

sample_df["W/F"] = pd.to_numeric(sample_df["W/F"].str.replace("★", "", regex=False).str.strip(), errors='coerce')
sample_df["SM"] = pd.to_numeric(sample_df["SM"].str.replace("★", "", regex=False).str.strip(), errors='coerce')
sample_df["IR"] = pd.to_numeric(sample_df["IR"].str.replace("★", "", regex=False).str.strip(), errors='coerce')

# Display the updated DataFrame
sample_df
```

```python
def money_to_numeric(series):
    return series.str.replace('[€KMB]', '', regex=True).replace({
        'K': '*1e3', 'M': '*1e6', 'B': '*1e9'}, regex=True).map(pd.eval).astype(float)

sample_df['Value'] = money_to_numeric(sample_df['Value'])
sample_df['Wage'] = money_to_numeric(sample_df['Wage'])

plt.figure(figsize=(10, 6))
plt.scatter(sample_df['Wage'], sample_df['Value'], alpha=0.6, edgecolor='k')

# Add titles and labels
plt.title('Players with High Value but Low Wages', fontsize=16)
plt.xlabel('Wage (€)', fontsize=14)
plt.ylabel('Value (€)', fontsize=14)

# Highlight players who are underpaid but have high value
underpaid_high_value = sample_df[(sample_df['Value'] > 1e7) & (sample_df['Wage'] < 1e5)]
plt.scatter(underpaid_high_value['Wage'], underpaid_high_value['Value'], color='red', label='Underpaid, High Value')


# Add legend
plt.legend()
plt.grid(True)
plt.show()
```
